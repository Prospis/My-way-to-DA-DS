```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("‚úÖ –ë–∏–±–ª–∏–æ—Ç–µ–∫–∏ –∑–∞–≥—Ä—É–∂–µ–Ω—ã")
```

```python
# ========================================
# –î–ê–¢–ê–°–ï–¢ 1: –ö–ª–∏–µ–Ω—Ç—ã (Customers)
# ========================================

customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 5, 6, 7, 8],
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank', 'Grace', 'Henry'],
    'city': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 
             'New York', 'Chicago', 'Los Angeles'],
    'age': [28, 35, 42, 29, 38, 45, 31, 52],
    'join_date': pd.to_datetime(['2023-01-15', '2023-02-20', '2023-01-10', 
                                  '2023-03-05', '2023-02-18', '2023-04-01',
                                  '2023-01-25', '2023-03-12'])
})

print("üìã CUSTOMERS:")
print(customers)
print()

# ========================================
# –î–ê–¢–ê–°–ï–¢ 2: –ó–∞–∫–∞–∑—ã (Orders)
# ========================================

orders = pd.DataFrame({
    'order_id': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'customer_id': [1, 2, 1, 3, 2, 4, 5, 1, 6, 7],  # –°–≤—è–∑—å —Å customers!
    'product': ['Laptop', 'Phone', 'Tablet', 'Laptop', 'Monitor', 
                'Keyboard', 'Mouse', 'Headphones', 'Laptop', 'Phone'],
    'amount': [1200, 800, 500, 1200, 300, 100, 50, 150, 1200, 800],
    'order_date': pd.to_datetime(['2023-02-01', '2023-03-15', '2023-02-20',
                                   '2023-04-10', '2023-03-25', '2023-05-05',
                                   '2023-04-15', '2023-05-20', '2023-06-01',
                                   '2023-05-10'])
})

print("üì¶ ORDERS:")
print(orders)
```

```python
merged = customers.merge(orders, on ='customer_id', how = 'inner')
print('INNER JOIN (–∫–ª–∏–µ–Ω—Ç—ã  –∏ –∏—Ö –∑–∞–∫–∞–∑—ã):')
print(merged)
print(f'–í—Å–µ–≥–æ —Å—Ç—Ä–æ–∫: {len(merged)}')
```

```python
merged_left = customers.merge(orders, on = 'customer_id', how='left')
print("üîó LEFT JOIN (–≤—Å–µ –∫–ª–∏–µ–Ω—Ç—ã, –¥–∞–∂–µ –±–µ–∑ –∑–∞–∫–∞–∑–æ–≤):")
print(merged_left)
print(f"\n–í—Å–µ–≥–æ —Å—Ç—Ä–æ–∫: {len(merged_left)}")

print("\nüîç –ü—Ä–æ–≤–µ—Ä–∫–∞: –µ—Å—Ç—å –ª–∏ Henry?")
henry = merged_left[merged_left['name'] == 'Henry']
print(henry)
```

```python
merged_left = customers.merge(orders, on = 'customer_id', how='left')
print("üîó LEFT JOIN (–≤—Å–µ –∫–ª–∏–µ–Ω—Ç—ã, –¥–∞–∂–µ –±–µ–∑ –∑–∞–∫–∞–∑–æ–≤):")
print(merged_left)
print(f"\n–í—Å–µ–≥–æ —Å—Ç—Ä–æ–∫: {len(merged_left)}")

print("\nüîç –ü—Ä–æ–≤–µ—Ä–∫–∞: –µ—Å—Ç—å –ª–∏ Henry?")
henry = merged_left[merged_left['name'] == 'Henry']
print(henry)
```

```python
customers_orders = customers.merge(orders, on = 'customer_id', how = 'left')

total_spent = customers_orders.groupby('name')['amount'].sum()

print("üí∞ –°–∫–æ–ª—å–∫–æ –∫–∞–∂–¥—ã–π –∫–ª–∏–µ–Ω—Ç –ø–æ—Ç—Ä–∞—Ç–∏–ª:")
print(total_spent.sort_values(ascending=False))
```

```python
print("\nüèÜ –¢–û–ü-3 –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ —Ç—Ä–∞—Ç–∞–º:")
print(total_spent.nlargest(3))
```

```python
pivot = customers_orders.pivot_table(
    values = 'amount',
    index = 'city',
    columns = 'product',
    aggfunc = 'sum',
    fill_value = 0
)

print("üìä –°–í–û–î–ù–ê–Ø –¢–ê–ë–õ–ò–¶–ê: –ü—Ä–æ–¥–∞–∂–∏ –ø–æ –≥–æ—Ä–æ–¥–∞–º –∏ –ø—Ä–æ–¥—É–∫—Ç–∞–º")
print(pivot)
```

```python
print("\nüíª Laptop: –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –≥–æ—Ä–æ–¥–∞–º")
print(pivot['Laptop'].sort_values(ascending=False))
```

```python
average_age_by_city = merged_left.groupby('city')['age'].mean().round(2)
print('C—Ä–µ–¥–Ω–∏–π –≤–æ–∑—Ä–∞—Å—Ç –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –≥–æ—Ä–æ–¥–∞–º')
print(average_age_by_city)
```

```python
count_of_orders_by_city = merged_left.pivot_table(
    values = 'order_id',
    index = 'city',
    columns = 'product',
    aggfunc = 'count',
    fill_value = '0'
)

print('–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤ –ø–æ –≥–æ—Ä–æ–¥–∞–º –∫–∞–∂–¥–æ–≥–æ –ø—Ä–æ–¥—É–∫—Ç–∞')
print(count_of_orders_by_city)
```

```python
last_customer_with_first_order = merged_left.groupby('name')['order_date'].min().sort_values(ascending=False)
print(last_customer_with_first_order)
```

```python
customer_spending = customers_orders.groupby('name')['amount'].sum().sort_values(ascending=False)

print("üí∞ –û–±—â–∏–µ —Ç—Ä–∞—Ç—ã –∫–ª–∏–µ–Ω—Ç–æ–≤:")
print(customer_spending)
```

```python
customer_value = customer_spending.apply(
    lambda x: 'Low Value' if x <= 500 
        else ('Medium Value' if x <= 1000 
              else 'High Value')
              )

print("\nüìä –ö–∞—Ç–µ–≥–æ—Ä–∏–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤:")
print(customer_value)

print("\nüìà –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º:")
print(customer_value.value_counts())
```

```python
city_stats = customers_orders.groupby('city').agg(
    total_customers = ('customer_id', 'count'),
    total_spent_by_all_customers = ('amount', 'sum'),
    total_average = ('amount', 'mean')
)

print(city_stats)
```

```python
idx = customers_orders.groupby('city')['amount'].idxmax()
top_client = customers_orders.loc[idx]

customer_report = pd.DataFrame({
    'Category': customer_value,
    'Total_Spent': customer_spending
    
})
final_table = top_client.merge(customer_report, on = 'name', how = 'left')



print(final_table[['city', 'name', 'Category', 'Total_Spent']])
```

```python
x = ['age', 'order_id', 'amount']
data = merged_left[x + ['product']].dropna()
sns.pairplot(data,hue = 'product')
plt.savefig('images/pairplot_3cols.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

```python
plt.figure(figsize=(12,6))
data_for_violinplot = merged_left.dropna()
sns.boxplot(
    x='product', 
    y='amount', 
    data=data_for_violinplot, 
    hue = 'product', 
    palette='Set2',
    legend=False)
plt.xticks(rotation=45)
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ü–µ–Ω –ø–æ –ø—Ä–æ–¥—É–∫—Ç–∞–º (Box Plot)', fontsize=14, fontweight='bold')
plt.ylabel('–¶–µ–Ω–∞ ($)')
plt.xlabel('–ü—Ä–æ–¥—É–∫—Ç')
plt.tight_layout()
plt.savefig('images/boxplot_products.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(12,6))
data_for_joinplot = merged_left.dropna()
sns.jointplot(
    x='age', 
    y='amount', 
    data=data_for_joinplot, 
    kind='reg'
    )
plt.xticks(rotation=45)
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–æ–∑—Ä–∞—Å—Ç–∞ –ø–æ —Ü–µ–Ω–∞–º (Join Plot)', fontsize=14, fontweight='bold')
plt.ylabel('–¶–µ–Ω–∞ ($)')
plt.xlabel('–í–æ–∑—Ä–∞—Å—Ç')
plt.tight_layout()
plt.savefig('images/joinplot_products.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(12,6))
data_for_catplot = merged_left.dropna()
sns.catplot(
    x='product', 
    y='amount', 
    data=data_for_catplot, 
    kind='bar',
    height=6,
    aspect=1.5
    )
plt.xticks(rotation=45)
plt.title('–°—Ä–µ–¥–Ω—è—è —Å—É–º–º–∞ –ø–æ –ø—Ä–æ–¥—É–∫—Ç—É', fontsize=14, fontweight='bold')
plt.ylabel('–¶–µ–Ω–∞ ($)')
plt.xlabel('–ü—Ä–æ–¥—É–∫—Ç')
plt.tight_layout()
plt.savefig('images/catplot_products.png', dpi=300, bbox_inches='tight')
plt.show()
```
